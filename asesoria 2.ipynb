1.- Dos equipos juegan una serie de juegos (4 de 7, el que gane 4 juegos gana). Cada equipo tiene un 50% de probabilidad de ganar cada juego (no empates). ¿Cuál es la probabilidad de que la serie se vaya a 7 juegos? (5/16)

```python
import numpy as np

n = 1000000
conteo = 0

for i in range(n):
    
    juegos = np.random.randint(0,2,7)   
    a = 0
    b = 0
    
    for j in juegos:
        if j == 0:
            a += 1
        else:
            b += 1
        
        if a == 4 or b == 4:
            break
    
    if a == 4 and b == 3:
        conteo += 1
    if b == 4 and a == 3:
        conteo += 1

prob = conteo / n * 100

print("Probabilidad estimada:", prob, "%")
```

2.- Una en mil personas tienen una enfermedad H, y la prueba de la enfermedad sale correcta 98% de las veces cuando la persona está enferma. Por otro lado, si la persona no está enferma la prueba tiene un error del 1%. Si alguien se hace una prueba y sale positiva, ¿cuál es la probabilidad de que tengan la enfermedad? (98/1097)



```python

n = 1000000  

positivos = 0              
enfermos_y_positivos = 0   

for i in range(n):

    enfermo = np.random.rand() < (1/1000)

    if enfermo:
        positivo = np.random.rand() < 0.98
    else:
        positivo = np.random.rand() < 0.01

    if positivo:
        positivos += 1

        if enfermo:
            enfermos_y_positivos += 1

prob = (enfermos_y_positivos / positivos) *100

print("Probabilidad aproximada:", prob ,"%")
```

3.- Tres amigos diferentes te dijeron que está lloviendo, y cada persona tiene 1/3 de probabilidad de mentir. Si el pronóstico del clima dijo que la probabilidad de lluvia era 25%, ¿cuál es la probabilidad de que sí esté lloviendo? (8/11)



```python

n = 1000000

casos = 0          
lluvia_casos = 0   

for i in range(n):

    lluvia = np.random.rand() < 0.25

    miente1 = np.random.rand() < (1/3)
    if lluvia:
        dice1 = not miente1
    else:
        dice1 = miente1

    miente2 = np.random.rand() < (1/3)
    if lluvia:
        dice2 = not miente2
    else:
        dice2 = miente2

    miente3 = np.random.rand() < (1/3)
    if lluvia:
        dice3 = not miente3
    else:
        dice3 = miente3

    if dice1 and dice2 and dice3:
        casos += 1

        if lluvia:
            lluvia_casos += 1

prob = (lluvia_casos / casos) * 100

print("Probabilidad aproximada:", prob, "%")
```
